```python
+producto_uno = [1, "Orden de tacos", 80]
producto_dos = [2, "Orden de enchiladas", 120]
producto_tres = [3, "Orden de gorditas", 90]
producto_cuatro = [4, "Orden de flautas", 130]
producto_cinco = [5, "Orden de enfrijoladas", 70]
producto_seis = [6, "Orden de sopes", 70]
producto_siete = [7, "Refresco", 20]
producto_ocho = [8, "Agua de horchata", 25]
producto_nueve = [9, "Agua de jamaica", 25]
producto_diez = [10, "Agua natural", 15]
alimentos = [producto_uno, producto_dos, producto_tres, producto_cuatro, producto_cinco, producto_seis, producto_siete, producto_ocho, producto_nueve, producto_diez]
compra = []
precio = []

def menu():
    print("La Fonda")
    print(f'{"ID":4}{"Alimento":25}{"Precio":10}')
    for num in range(10):
        print(f"{alimentos[num][0]:<6}{alimentos[num][1]:<25} ${float(alimentos[num][2]):<10}")
    venta()

def venta():
    while True:
        try:
            
            ID = input("Ingrese la ID del alimento que desea comprar (Favor de finalizar su compra con *): ")

            if ID != "*":
                ID = int(ID)

                if ID < 1 or ID > 10:
                    print("Producto no identificado")

                elif ID == 1:
                    compra.append(producto_uno[0])
                    precio.append(producto_uno[2])

                elif ID == 2:
                    compra.append(producto_dos[0])
                    precio.append(producto_dos[2])

                elif ID == 3:
                    compra.append(producto_tres[0])
                    precio.append(producto_tres[2])

                elif ID == 4:
                    compra.append(producto_cuatro[0])
                    precio.append(producto_cuatro[2])
                
                elif ID == 5:
                    compra.append(producto_cinco[0])
                    precio.append(producto_cinco[2])

                elif ID == 6:
                    compra.append(producto_seis[0])
                    precio.append(producto_seis[2])

                elif ID == 7:
                    compra.append(producto_siete[0])
                    precio.append(producto_siete[2])

                elif ID == 8:
                    compra.append(producto_ocho[0])
                    precio.append(producto_ocho[2])

                elif ID == 9:
                    compra.append(producto_nueve[0])
                    precio.append(producto_nueve[2])
                
                elif ID == 10:
                    compra.append(producto_diez[0])
                    precio.append(producto_diez[2])

            elif ID == "*":
                if len(compra) == 0:
                    print("Gracias por adquirir nada, vuelva pronto")
                    break

                print(f'{"Alimento":25}{"Precio":10}')
                for elemento in compra:
                    elemento = int(elemento)
                    acum = 0
                    
                    for dato in precio:
                        acum = acum + dato
                    
                    IVA = acum * 0.16

                    print(f"{alimentos[elemento - 1][1]:<23} $ {alimentos[elemento - 1][2]:<6}")
                print(f'{"IVA":<5} $ {round(IVA, 3)}')
                print(f'{"El total a pagar es"} $ {acum + IVA}')
                print("¡Espero lo disfrute!")
                break

        except ValueError:
            print("Favor de ingresar un dato correcto")
            continue

menu()
```
