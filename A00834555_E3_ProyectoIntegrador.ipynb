# Codigo Fuente

```python
lista_empleados = []
peso_bajo = []
peso_normal = []
sobrepeso = []
obesidad = []
diabs = []
hiper = []
canc = []
cora = []
tabaq = []
vacu = []

nomb = "empresa.txt"

def leer_archivo(titulo):
    
    base_empleados = []

    with open(titulo, "r") as archivo:
        contenido = archivo.readlines()

        for dato in contenido:
            renglon = dato.split(" , ")
            
            if len(renglon) == 11:
                renglon[11] = renglon[11][:-1]
            
            elif len(renglon) == 13: 
                renglon[11] = renglon[11][:-1]
                renglon[13] = renglon[13][:-1]

            base_empleados.append(renglon)

        print(contenido)

def grabar_archivo(titulo):
    with open(titulo, "w") as archivo:
        
        for empleado in lista_empleados:
            empleado[0] = str(empleado[0]) # ID
            empleado[2] = str(empleado[2]) # Edad
            empleado[3] = str(empleado[3]) # Peso
            empleado[4] = str(empleado[4]) # Estatura
            empleado[11] = str(empleado[11]) # Departamento

            if len(empleado) >= 13:
                empleado[13] = str(empleado[13]) # IMC            
            
            texto = " , ".join(empleado)
            archivo.write(texto + "\n")

def alta_empleado():

    leer_archivo(nomb)

    global lista_empleados
    global empresa

    empleado = []

    try:

        id = int(input("Teclear la ID del empleado: "))
        if type(id) != int:
            print("Favor de insertar un dato válido")
            alta_empleado()

        nombre = str(input("Teclear el nombre del empleado: "))
        if type(nombre) != str:
            print("Favor de insertar un dato válido")
            alta_empleado()

        edad = int(input("Teclear la edad del empleado: "))
        if type(edad) != int:
            print("Favor de insertar un dato válido")
            alta_empleado()

        peso = float(input("Teclear el peso del empleado (kg): "))
        if type(peso) != int and type(peso) != float:
            print("Favor de insertar un dato válido")
            alta_empleado()

        estatura = float(input("Teclear la estatura del empleado (mts): "))
        if type(estatura) != int and type(estatura) != float:
            print("Favor de insertar un dato válido")
            alta_empleado()

        diabetes = int(input("¿Tiene diabetes el empleado? 1.- Sí 0.- No : "))
        if diabetes == 1:
            diabetes = "diabetes"
            diabs.append(id)
        elif diabetes == 0:
            diabetes = "_" * 6
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        hipertension = int(input("¿Tiene hipertensión el empleado? 1.- Sí 0.- No : "))
        if hipertension == 1:
            hipertension = "hipertension"
            hiper.append(id)
        
        elif hipertension == 0:
            hipertension = "_" * 10
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        corazon = int(input("¿Tiene alguna enfermedad del corazón el empleado? 1.- Sí 0.- No : "))
        if corazon == 1:
            corazon = "corazon"
            cora.append(id)
        elif corazon == 0:
            corazon = "_" * 5
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        cancer = int(input("¿Tiene cancer el empleado? 1.- Sí 0.- No : "))
        if cancer == 1:
            cancer = "cancer"
            canc.append(id)
        elif cancer == 0:
            cancer = "_" * 4
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        tabaco = int(input("¿Presenta tabaquismo el empleado? 1.- Sí 0.-No : "))
        if tabaco == 1:
            tabaco = "tabaquismo"
            tabaq.append(id)

        elif tabaco == 0:
            tabaco = "_" * 8
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        vacuna = int(input("¿Ya se ha vacunado el empleado? 1.- Sí 0.- No : "))
        if vacuna == 1:
            vacuna = "Con vacuna"
            vacu.append(id)
        elif vacuna == 0:
            vacuna = "Sin vacuna"
        else:
            print("Favor de insertar un dato válido")
            alta_empleado()

        departamento = int(input("Favor de registrar el departamento al que pertenece el empleado: "))

        empleado = [id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaco, vacuna, departamento]
        print(empleado)

        lista_empleados.append(empleado)

        grabar_archivo(nomb)

    except ValueError:
        print("Ingrese un dato válido porfa ;^;")

def imc_empleado():
    
    leer_archivo(nomb)

    global lista_empleados

    imc_id = int(input("Ingrese la ID del empleado: "))

    if imc_id > len(lista_empleados):
        print("El empleado no existe")
    
    else:
        imc = float(lista_empleados[imc_id - 1][3]) / (float(lista_empleados[imc_id - 1][4]) ** 2)
    
        print(round(imc, 2))

        if imc < 18.5:
            print("Bajo Peso")
            lista_empleados[imc_id - 1].append("Bajo Peso")
            peso_bajo.append(imc_id)
            if imc < 16:
                print("Delgadez Severa")
            elif imc == 16 or imc < 17:
                print("Delgadez Moderada")
            else:
                print("Delgadez Aceptable")
        
        elif imc == 18.5 or imc < 24.5:
            print("Peso Normal")
            lista_empleados[imc_id - 1].append("Peso Normal")    
            peso_normal.append(imc_id)

        elif imc == 24.5 or imc < 30:
            print("Sobrepeso")
            lista_empleados[imc_id - 1].append("Sobrepeso")
            sobrepeso.append(imc_id)

        elif imc == 30 or imc > 30:
            print("Obeso")
            lista_empleados[imc_id - 1].append("Obesidad")
            obesidad.append(imc_id)
            if imc == 30 or imc < 35:
                print("Tipo I (Riesgo Moderado)")
            elif imc == 35 or imc < 40:
                print("Tipo II (Riesgo Severo)")
            else:
                print("Tipo III (Riesgo Muy Severo")

        imc = str(round(imc, 2))
    
        lista_empleados[imc_id - 1].append(imc)

        grabar_archivo(nomb)

def cambiar_info():
    
    leer_archivo(nomb)

    global lista_empleados

    lista_temp = []

    num_emp = int(input("Ingrese la ID del empleado que desea cambiar información: "))

    n = 0
    while n < len(lista_empleados[num_emp - 1]):
        lista_temp.append(lista_empleados[num_emp - 1][n])
        n = n +1

    print("""Datos a cambiar: \n
    1. Nombre
    2. Edad
    3. Peso
    4. Estatura
    5. Diabetes
    6. Hipertension
    7. Corazon
    8. Cancer
    9. Tabaco
    10. Vacuna
    11. Departamento""")

    num_dato = int(input("Ingrese el dato que desea cambiar: "))

    if num_dato < 1 or num_dato > 11:
        print("Ingrese un dato valido")
        cambiar_info()
    
    else:
        nuevo_dato = input("Inserte el nuevo dato: ")
        nuevo_dato = str(nuevo_dato)

    lista_temp.pop(num_dato)
    lista_temp.insert(num_dato, nuevo_dato)

    print("Nueva Lista: ")
    print(lista_temp)

    cambiar = int(input("¿Desea cambiar los datos? 1.- Sí 0.- No"))

    if cambiar == 1:
        lista_empleados[num_emp - 1][num_dato] = nuevo_dato
    elif cambiar == 0:
        print("Cambio no Efectuado")
    else:
        print("Error al ingresar los datos, favor de repetir el proceso")
        menu()
    
    grabar_archivo(nomb)

def consulta_empleado():
    
    leer_archivo(nomb)

    global lista_empleados

    consulta = int(input("Ingrese la ID del empleado que desea ver: "))
  
    if consulta < len(lista_empleados) or consulta == len(lista_empleados):
        print(f"Empleado {consulta} \n {lista_empleados[consulta - 1]}")
  
    else:
        print("El empleado no existe")
    
def reporte_empledos():
    
    global lista_empleados
    
    for empleado in lista_empleados:
        if len(empleado) != 14:
            print("Favor de calcular todos los IMC")
            main()
        
        else:
            print("")

    print(f'{"ID":5}{"Nombre":9}{"Edad":5}{"Peso":6}{"Estatura":10}{"Diabetes":10}{"Hipertension":13}{"Corazon":8}{"Cancer":7}{"Tabaco":11}{"Vacuna":11}{"Departamento":13}{"Estatus":21}{"IMC":6}')
    for empleado in lista_empleados:
        print(f'{empleado[0]:<6}{empleado[1]:<11}{empleado[2]:<6}{empleado[3]:<7}{empleado[4]:<11}{empleado[5]:<10}{empleado[6]:<13}{empleado[7]:<8}{empleado[8]:<7}{empleado[9]:<11}{empleado[10]:11}{empleado[11]:<16}{empleado[12]:<20}{empleado[13]:<7}')

    print(f'{"Diagnostico":12}{"Cantidad":9}{"Porcentaje":10}')
    print("Peso bajo", " " * 7,len(peso_bajo), " " * 6, (len(peso_bajo) * 100)/len(lista_empleados), "%")
    print("Peso normal", " " * 4, len(peso_normal), " " * 4, (len(peso_normal) * 100)/len(lista_empleados), "%")
    print("Sobrepeso", " " * 6, len(sobrepeso), " " * 5, (len(sobrepeso) * 100)/len(lista_empleados), "%")
    print("Obesidad", " " * 7, len(obesidad), " " * 5, (len(obesidad) * 100)/len(lista_empleados), "%")

    print(f'{"Enfermedad":12}{"Cantidad":9}{"Porcentaje":10}')
    print("Diabetes", " " * 7,len(diabs), " " * 6, (len(diabs) * 100)/len(lista_empleados), "%")
    print("Hipertension", " " * 4, len(hiper), " " * 4, (len(hiper) * 100)/len(lista_empleados), "%")
    print("Corazon", " " * 6, len(cora), " " * 5, (len(cora) * 100)/len(lista_empleados), "%")
    print("Cancer", " " * 7, len(canc), " " * 5, (len(canc) * 100)/len(lista_empleados), "%")
    print("Tabaco", " " * 7, len(tabaq), " " * 5, (len(tabaq) * 100)/len(lista_empleados), "%")
    print("Vacuna", " " * 7, len(vacu), " " * 5, (len(vacu) * 100)/len(lista_empleados), "%") 


def menu():
    print("Sistema de Salud y Bienestar del Empleado")
    print("1. Alta de empleado")
    print("2. Calcular imc de empleados")
    print("3. Cambiar la información del empleado")
    print("4. Consulta un empleado")
    print("5. Reporte de todos los empleados")
    print("6. Salir")

    option = int(input("Seleccione una opción: "))

    return option

def main():

    global lista_empleados

    continuar = True

    while continuar:
        option = menu()
   
        if option < 1 or option >6:
            print("La opción tecleada no existe")
            continue

        elif option == 1:
            alta_empleado()

        elif option == 2:
            imc_empleado()
     
        elif option == 3:
            cambiar_info()
   
        elif option == 4:
            consulta_empleado()
   
        elif option == 5:
            reporte_empledos()
   
        else:
            print("Gracias por su atención, bonito día")
            continuar = False

main()
```
